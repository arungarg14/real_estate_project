```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import matplotlib 
matplotlib.rcParams['figure.figsize'] = (20,10)
```

```python
df1 = pd.read_csv('Bengaluru_House_Data.csv')
df1.head()
```

```python
df1.shape
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
#Assuming, these factors ['area_type','society','balcony','availability'] will not contribute in prediction.

df2 = df1.drop(['area_type','society','balcony','availability'], axis='columns')
df2.shape
```

```python
df2.head()
```

```python
#Total no of null values in each columns...

df2.isnull().sum()
```

```python
#Drop all the na values because na is small comparable to dataframe

df3 = df2.dropna()
df3.isnull().sum()
df3.shape
```

```python
#Data Cleaning part noow.. 

df3['size'].unique()
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3.head()    #Now, You have created the new columns consist of bhk
```

```python
df3['total_sqft'].unique()     
```

```python
#Now, we will convert range into medium value. First let see all this type of value.

def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

df3[~df3['total_sqft'].apply(is_float)]  #So, total 190 values are of that..
```

```python
#Convert range into average value...

def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2 
    try:
        return float(x)
    except:
        return None
    
```

```python
#Let's apply the function now.. 

df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4['total_sqft'].unique()
```

```python
df4.loc[13299]  #To see particular dataframe
```

```python
#Feature Engineering .......

df5 = df4.copy()
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
```

```python
#Location exploring...

len(df5.location.unique())   #It will cause dimensionality problem. So, we will decrease the dimensions.
```

```python
#Remove extra spaces between them (Location column)...

df5.location = df5.location.apply(lambda x: x.strip())
```

```python
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending=False)

location_stats.head(20)
```

```python
len(df5.location.unique())   #There is as much value in the location which is unique, which is very large (Dimensionality curse). so, we will 
                            #decrease no of loaction. Location which is less in number, assign them 'other' 
```

```python
location_stats_lessthan_10 = location_stats[location_stats<=10]
location_stats_lessthan_10
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_lessthan_10 else x) 
len(df5.location.unique())  #Now, the unique value has decreased..
```

```python
#Outliers are data errors. Sometimes, they are extreme variations. It makes sense to remove them..

#If we assume that typical flat has minimun 300 sqft in 1 bhk. Then, 

df5[(df5.total_sqft/df5.bhk)<300].head()    #Now, we have to remove these columns.
```

```python
df6 = df5[~((df5.total_sqft/df5.bhk)<300)]
df6.shape
```

```python
#Now, we will filter out values in price_per_sqft which is beyond the between mean and 1 std. 

def remove_pps_outlier(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
        
    return df_out

df7 = remove_pps_outlier(df6)
df7.shape
```

```python
#make sure that price of 3 BHK is must be greater than price of 2 BHK

def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
#Make sure that no of bathroom must not be greater than bhk+2 

df9 = df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
df10 = df9.drop(['size','price_per_sqft'], axis='columns')
df10.head()
```

```python
#Now, we will convert location in to ONE HOT ENCODING.

dummies = pd.get_dummies(df10.location)
dummies = dummies.drop('other', axis='columns')  #avoid dummy trap

dummies
```

```python
#Now, let's concatenate dummies and original dataframe.. 

df11 = pd.concat([df10,dummies], axis='columns')
df11.head()     

#Finally, we will drop the location column
```

```python
df12 = df11.drop('location', axis='columns')
df12.head()
```

```python
# Now, We can start with model building.. 
```

```python
x = df12.drop('price', axis='columns')
x.head()
```

```python
y = df12.price
y.head()
```

```python
#Splitting the dataset.. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
#Training the model...

from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
#Check the accuracy.. 

linear_reg.score(X_test, y_test)
```

```python
#Hurray! It is good. Now, time to do K-fold cross Validation..

from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
def predict_price(x,location,sqft,bath,bhk):  
    location_index = np.where(x.columns==location)[0][0]
    x = np.zeros(len(x.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if location_index >= 0:
        x[location_index] = 1

    return linear_reg.predict([x])
```

```python
predict_price(x,'1st Phase JP Nagar',1000, 2, 2)
```
